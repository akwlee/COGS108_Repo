This file contains our data cleaning for business.csv. We set our search to find restaurants with more than 50 reviews and a rating greater than or equal to 4.0 stars.

```python
import pandas as pd
import numpy as np
from tqdm import tqdm

businesses = pd.read_csv('yelp_business.csv')
```

```python
average_review_count = businesses['review_count'].mean()
# print(average_review_count)
top_bs = businesses[(businesses['stars'] == 5.0)]
top_bs.reset_index(inplace=True)
del top_bs['index']
top_bs
```

```python
# To limit our search to just Restaurants:
# Loop through each row and get the categories column value (a string)
# For each category, check if the word 'Restaurant' is in there and if not, drop that entire line in top_bs
count = 0;
with tqdm(total=27540) as pbar:
    for ind, row in top_bs.iterrows():
        if (('Restaurants' in row['categories']) or ('Food' in row['categories'])):
            count = count + 1
            pbar.update(1)
        else:
            count = count
            top_bs.drop(ind, inplace = True)
            pbar.update(1)
```

```python
# Drop unnecessary columns neighborhood and postal code.
top_bs.drop(['neighborhood','postal_code'], axis = 1, inplace = True)
```

```python
top_bs.drop(top_bs[top_bs['is_open'] == 0].index, inplace = True)
```

```python
top_bs.to_csv('updated_businesses.csv')
```
